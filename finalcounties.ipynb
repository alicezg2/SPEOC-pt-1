```python
import geopandas as gpd
import matplotlib.pyplot as pet
import pandas as pd
import os 

#full path
#os.chdir('/Users/mariadubasov/Downloads')

#fp = r"historical_county_shapefiles"

map_df = gpd.read_file("../../Data/StateShapeFiles/historicalcounties")

#map_df = gpd.read_file(fp)
map_df.rename(columns = {'NHGISNAM':'county'}, inplace = True)
map_df.rename(columns = {'STATENAM':'state'}, inplace = True)

map_df.head()
#map_df.plot()


file = "https://raw.githubusercontent.com/alicezg2/SPEOC-pt-1/main/Data/AssetGeography/county_debt_total.csv"
df = pd.read_csv(file, header=0)
df['county'] = df['county'].str.replace(" County","")


states = {"Connecticut": "CT","Delaware": "DE","Georgia": "GA", "Maryland": "MD", "Massachusetts": "MA", 
                      "New Hampshire": "NH", "New Jersey": "NJ", "New York": "NY","North Carolina": "NC",
                      "Pennsylvania": "PA", "Rhode Island": "RI", "South Carolina": "SC", "Virginia": "VA", }
    
map_df['state'] = map_df['state'].replace(states)

merged = pd.merge(map_df, df, on=['county', 'state'], how='left')


merged['6p_total'] = merged['6p_total'].fillna(0)

variable = '6p_total'
vmin, vmax = 0.01, 60000
cmap = pet.get_cmap('YlOrRd').copy()
cmap.set_under('white')
cmap.set_over('blue')
#fig, ax = pet.subplots(1, figsize=(10, 6))
fig, ax = pet.subplots(1, figsize=(20, 12))
#merged.plot(column=variable, cmap= cmap, linewidth=0.8, ax=ax, edgecolor='0.8', vmin=vmin, vmax = vmax)

ax.axis('off')

#sortlist = merged['6p_total'].sort_values()
#print(sortlist)

va_df = merged[merged['state'] == 'VA']
vamax = va_df['6p_total'].max()

merged[merged['state'] == 'VA'].apply(lambda merged: ax.annotate(text=merged['county'], fontsize = 5, xy=merged['geometry'].centroid.coords[0], ha='center'), axis=1)
merged[merged['state'] == 'VA'].plot(column=variable, cmap= cmap, linewidth=0.8, ax=ax, edgecolor='0.8', vmin=vmin, vmax = vamax)

sm = pet.cm.ScalarMappable(cmap='YlOrRd', norm=pet.Normalize(vmin=vmin, vmax=vamax))
sm._A = []
cbar = fig.colorbar(sm)
```
